```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Age': [25, 27, np.nan, 35, 29],
    'Salary': [50000, 54000, 58000, None, 62000]
})

# Impute with mean & median
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

print("# Missing Value Imputation")
print(df)
print()
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Pink', 'Red']
})
df_encoded = pd.get_dummies(df, columns=['Color'])
print("# Encoding Categorical Variables")
print(df_encoded)
print()
from sklearn.preprocessing import StandardScaler

df = pd.DataFrame({
    'Income': [30000, 50000, 70000],
    'Age': [25, 37, 45]
})

scaler = StandardScaler()
scaled_df = scaler.fit_transform(df)
print("# Feature Scaling (Standardization)")
print(scaled_df)
print()
df = pd.DataFrame({
    'Income': [30000, 50000, 100000]
})

df['Log_Income'] = np.log(df['Income'])
print("# Feature Transformation (Log)")
print(df)
print()
df = pd.DataFrame({
    'Total_Sales': [100, 200, 300],
    'Total_Orders': [10, 20, 30]
})

df['Avg_Order_Value'] = df['Total_Sales'] / df['Total_Orders']
print("# Feature Creation")
print(df)
print()
df = pd.DataFrame({'Age': [15, 20, 30, 45, 60]})

df['Age_Group'] = pd.cut(df['Age'], bins=[0, 18, 35, 60], labels=['Teen', 'Adult', 'Senior'])
print("# Discretization / Binning")
print(df)
print()
df = pd.DataFrame({
    'Length': [5, 12, 15],
    'Width': [2, 4, 6]
})

df['Area'] = df['Length'] * df['Width']
print("# Interaction Features")
print(df)
print()
from sklearn.preprocessing import PolynomialFeatures

df = pd.DataFrame({
    'x': [1, 3, 4]
})

poly = PolynomialFeatures(degree=2, include_bias=False)
print("# Polynomial Features")
print(poly.fit_transform(df))
print()
df = pd.DataFrame({'date': pd.to_datetime(['2025-08-05', '2025-08-07'])})

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['weekday'] = df['date'].dt.weekday
print("# Datetime Features")
print(df)
print()
from sklearn.feature_extraction.text import TfidfVectorizer

docs = ['Machine learning is Fun', 'Deep learning is Powerful']

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)
print("# Text Feature Engineering (TF-IDF)")
print(X.toarray())
print()
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [2, 4, 6],
    'C': [3, 6, 9]
})
corr = df.corr()
print("# Feature Selection Techniques (Correlation Heatmap)")
sns.heatmap(corr, annot=True)
plt.show()
```
